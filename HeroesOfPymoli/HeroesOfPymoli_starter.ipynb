
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

```python

male = 0
female = 0
other = 0

index = 0

pricecol = game_data["Price"]
maleprice = 0
femaleprice = 0
otherprice = 0

uniquemales = []
uniquefemales = []
uniqueothers = []

usernames = game_data["SN"]

for i in game_data["Gender"]:
    if i == 'Male': 
        male+=1
        maleprice = maleprice + pricecol[index]
        if uniquemales.count(usernames[index]) == 0:
            uniquemales.append(usernames[index])
    elif i == 'Female': 
        female+=1
        femaleprice = femaleprice + pricecol[index]
        if uniquefemales.count(usernames[index]) == 0:
            uniquefemales.append(usernames[index])
    elif i == "Other / Non-Disclosed": 
        other+=1
        otherprice = otherprice + pricecol[index]
        if uniqueothers.count(usernames[index]) == 0:
            uniqueothers.append(usernames[index])
    index+=1
    
    

genders = {'Male' : male, 'Female' : female, 'Other' : other}

maleprice = round(maleprice,2)
femaleprice = round(femaleprice,2)
otherprice = round(otherprice,2)

malepricefinal = "$" + str(maleprice)
femalepricefinal = "$" + str(femaleprice)
otherpricefinal = "$" + str(otherprice)


maleaverage = round((maleprice / male),2)
femaleaverage = round((femaleprice/ female),2)
otheraverage = round((otherprice/ other),2)

maleaverage = "$" + str(maleaverage)
femaleaverage = "$" + str(femaleaverage)
otheraverage = "$" + str(otheraverage)


MalePerPerson = round((maleprice / len(uniquemales)),2)
MalePerPerson = "$" + str(MalePerPerson)

FemalePerPerson = round((femaleprice / len(uniquefemales)),2)
FemalePerPerson = "$" + str(FemalePerPerson)

OtherPerPerson = round((otherprice / len(uniqueothers)),2)
OtherPerPerson = "$" + str(OtherPerPerson)





GenderPurchaseDF = {'Purchase count' : [male,female,other], 
                    'Average Purchase Price' : [maleaverage,femaleaverage,otheraverage],
                   'Total purchase value' : [malepricefinal,femalepricefinal,otherpricefinal],
                    'Average total purchase per person' : [MalePerPerson,FemalePerPerson,OtherPerPerson]
                   }
GenderPurchaseDF = pd.DataFrame(data=GenderPurchaseDF, index = ['Male','Female','Other'])
 
GenderPurchaseDF
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemid_data = game_data["Item ID"]
itemname_data = game_data["Item Name"]
price_data = game_data["Price"]


#dicitonary of itemname : purchase count
purchase_count = {}
#dicitonary of itemname : price
price_tracker = {}
id_tracker = {}

totalpurchase = []
index = 0
prices = []
for key in itemname_data:
    
    #print(key)
    id_tracker.setdefault(key)
    purchase_count.setdefault(key)
    price_tracker.setdefault(key)
    id_tracker.setdefault(key)
    if purchase_count[key] == None:
        purchase_count[key] = 1
    else:
        value = purchase_count[key] + 1
        purchase_count[key] = value
     
    if price_tracker[key] == None:
        price_tracker[key] = price_data[index]
    if id_tracker[key] == None:
        id_tracker[key] = itemid_data[index]
    index+=1


count = []
names = []
prices = []  
itemids = []
for key, value in purchase_count.items():
    names.append(key)
    count.append(value)
    
for key, value in price_tracker.items():
    prices.append(value)

for key, value in id_tracker.items():
    itemids.append(value)

index2 = 0

while index2 < len(prices): 
    totalpurchase.append(prices[index2] * count[index2])
    prices[index2] = "$" + str(prices[index2])
    totalpurchase[index2] = "$" + str(round((totalpurchase[index2]),2))
    index2+=1
    
    
totalpurchase = pd.Series(data = totalpurchase)    
count = pd.Series(data = count)
names = pd.Series(data = names)
prices = pd.Series(data = prices)
itemids = pd.Series(data = itemids)
predf = {'Item ID': itemids, 'Item Name' : names, 'Purchase Count' : count, 'Price' : prices, 
         'Total Purchase value' : totalpurchase}
df1 = pd.DataFrame(data = predf)




df1 = df1.sort_values("Purchase Count", ascending=False)

df1.head()
 
```
